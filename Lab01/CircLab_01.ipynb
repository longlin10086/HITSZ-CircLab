### u,i 曲线图绘制

```python
# f=f0  f<f0  f>f0
u = [1.5, 1.5, 1.5] # V
I = [1.6, 1.4, 1.6] # A
fi = [0, 56, -75] # 相位差
f_u = [10.7, 3, 19] # kHz
```

```python
T = []
X = []
Omega = []
for i in range(0, len(f_u)):
    T.append(round(1/f_u[i], 2)) # ms
    Omega.append(round(2*np.pi*f_u[i], 2))
```

```python
print(f"period(ms): {T}")
print(Omega)
```

```python
s_u = []
s_i = []
for i in range(0, len(T)):
    X = np.arange(0.0, 3*T[i], 0.001)
    s_u.append(u[i]*np.sin(Omega[i]*X))
    s_i.append(I[i]*np.sin(Omega[i]*X+fi[i]))
    fig, ax = plt.subplots()
    ax.plot(X, s_u[i])
    ax.plot(X, s_i[i])
    ax.set_xlabel(f'T(ms): {T[i]}, phase diff: {fi[i]}°, u_max(V): {u[i]}, i_max(A): {I[i]}')
    ax.set_ylabel(r'u and i')
    match i:
        case 0:
            ax.set_title('f=f0')
        case 1:
            ax.set_title('f>f0')
        case 2:
            ax.set_title('f<f0')
    plt.show()
```
